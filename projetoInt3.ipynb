```python

### função de filtro que separa as palavras e remove conectores
def filtered(search):
   
    ## recebe from textblob biblioteca para reformulação da palavras
     #############################
    #lista
    lem = []
    ##separa as palavras e reformula ela
    for i in search.split():
        word1 = Word(i).lemmatize("n")
        word2 = Word(word1).lemmatize("v")
        word3 = Word(word2).lemmatize("a")
        lem.append(Word(word3).lemmatize())
    #############################
    ##usa spacy para retirar os conectores
     ##filtro para ingles
    nlp = English()
    ##lista para guardar o que foi filtrado
    filtered_sentence =[] 
    ##remove os conectores
    for word in lem:
        lexeme = nlp.vocab[word]
        if lexeme.is_stop == False:
            filtered_sentence.append(word) 
    ##retorna as palavras filtradas
    return filtered_sentence
```

```python
##Classifica as palavras com base na biblioteca do Spacy
##função
def ner(search):
    ##lista vazia
    entities = []
    ##carrega a biblioteca do NER do spacy***ingles
    NER = spacy.load("en_core_web_sm")
    ##repetição que interage com cada palvra da função de filtragem 
    for i in filtered(search):
        ##classifica a palavra com a biblioteca ner
        text1= NER(i)
        ##interage com cada palvra e guarda a classicificação na lista entities
        for word in text1.ents:
             entities.append(word.label_)
    ##retorna a lista                
    return entities  
```

```python
## função
def historico(search):
    ## pega a data de hoje
    now = datetime.now()
    ## mapeia a data
    d1 = now.strftime("%Y-%m-%d %H:%M:%S")
    ## codigo sql para adicionar no bd
    insert = "INSERT INTO hist (hist_texto, hist_data) VALUES (%s, %s)"
    ## variavel contendo o texto da pesquisa e a data de agora
    val = (search,d1)
    ## armazena o bd em uma variavel
    bd=historicoBD()
    ##adiciona um cursor localmente
    cursor = bd.cursor()
    ##adiciona no bd
    cursor.execute(insert, val)
    bd.commit()
```

```python
##verificador de texto
def verificadorTexto(search):
    ## mostra o texto antes das palavras corrigidas
    print("Voce quis dizer",end = ' ')
    ##interage com cada palavvra que foi separada da função filtro
    for word in filtered(search):
        ##detecta o idioma
        lang = detect(search)
        ## pega o idioma e corrigi de acordo com ele
        ## se ele for diferente de portugues, ingles e espanhol ele corrige em portugues
        if str(lang) !='pt' or  str(lang) !='en' or str(lang) !='es':
            ##mostra a palavra corrigida em portugues para idiomas diferentes de pt, en e es
            print(str(SpellChecker(language='pt').correction(word)),end = ' ')  
        else:
            print(str(SpellChecker(language=str(lang)).correction(word)),end = ' ')
    return
  
```

```python
search = 'action'
##função da pesquisa relacionadas
def pesquisaRelacionadas(search):
    ##codigo sql para chamar o historico
    sql = 'SELECT * FROM hist '
    ## le a tabela do historico  de pesquisa
    df = pd.read_sql(sql,historicoBD())
    ###variavel vazia onde ira conter as palavras no historico separadas
    df2=[]
    ##repetição para interagir com cada linha do historico e usar a função do filtro que radicalizar as palavras
    for index, row in df.iterrows():
        word=np.array(filtered(row['hist_texto']))
        df2.append(word)
    #### passa df2 para tipo dataframe
    df2=pd.DataFrame(df2)
    ###lista de palavras
    listaDePalavra = []
    ##tem que transformar cada posição do df2 em string
    for i in range(0, len(df2)):
        listaDePalavra.append([str(df2.values[i,j]) for j in range(0, len(df. columns)+1)])
    ###metodo Regra de Associação
    rules = apriori(transactions = listaDePalavra, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
    ###resultado
    results = list(rules)
    ##função que retorna os resultados em dicionario
    def inspect(results):
        lhs         = [tuple(result[2][0][0])[0] for result in results]
        rhs         = [tuple(result[2][0][1])[0] for result in results]
        supports    = [result[1] for result in results]
        return list(zip(lhs, rhs, supports))
    ###transforma em dataframe
    resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Palavra 1', 'Palavra 2', 'Support'])
    ##transforma em dataframe e mostra por ordem de maior suporte
    resultados=resultsinDataFrame.nlargest(n = 10, columns = 'Support')
    ###localiza as linhas que contem as palavras pesquisadas
    resultados=resultados[resultados['Palavra 1'].str.contains(search) | resultados['Palavra 2'].str.contains(search)]
    ###mostra os resultados
    return resultados 

pesquisaRelacionadas(search)
```
